```python
# Task 1: Explore the Menu

# given data (do not modify)
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# Part 1: Print menu grouped by category

categories = ["Starters", "Mains", "Desserts"]

for cat in categories:
    print(f"===== {cat} =====")

    # loop through menu items
    for item in menu:
        if menu[item]["category"] == cat:

            price = menu[item]["price"]
            available = menu[item]["available"]

            # checking availability
            if available:
                status = "Available"
            else:
                status = "Unavailable"

            print(f"{item:<15} ₹{price:.2f}   [{status}]")

    print()  # blank line after each category


# Part 2: Calculations

# total number of items
total_items = len(menu)
print("Total items on menu:", total_items)

# total available items
available_count = 0
for item in menu:
    if menu[item]["available"]:
        available_count += 1

print("Available items:", available_count)


# finding most expensive item
max_price = 0
max_item = ""

for item in menu:
    price = menu[item]["price"]

    if price > max_price:
        max_price = price
        max_item = item

print("Most expensive item:", max_item, "- ₹" + str(max_price))


# items under ₹150
print("\nItems under ₹150:")

for item in menu:
    price = menu[item]["price"]

    if price < 150:
        print(f"{item} - ₹{price}")
```

```python
# Task 2: Cart Operations

# starting with empty cart
cart = []

# function to add item to cart

def add_to_cart(item_name, qty):

    # check if item exists in menu
    if item_name not in menu:
        print(f"{item_name} is not available in menu.")
        return

    # check if item is available
    if not menu[item_name]["available"]:
        print(f"{item_name} is currently unavailable.")
        return

    # check if item already in cart
    found = False

    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += qty
            found = True
            break

    # if not found, add new entry
    if not found:
        cart.append({
            "item": item_name,
            "quantity": qty,
            "price": menu[item_name]["price"]
        })


# function to remove item
def remove_from_cart(item_name):

    found = False

    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            found = True
            break

    if not found:
        print(f"{item_name} not found in cart.")


# function to display cart
def show_cart():
    print("\nCurrent Cart:")

    if len(cart) == 0:
        print("Cart is empty.")
        return

    for entry in cart:
        print(f"{entry['item']} x{entry['quantity']}")


# Simulation Steps

# 1. Add Paneer Tikka ×2
add_to_cart("Paneer Tikka", 2)
show_cart()

# 2. Add Gulab Jamun ×1
add_to_cart("Gulab Jamun", 1)
show_cart()

# 3. Add Paneer Tikka ×1 (should update)
add_to_cart("Paneer Tikka", 1)
show_cart()

# 4. Try adding non-existing item
add_to_cart("Mystery Burger", 1)
show_cart()

# 5. Try adding unavailable item
add_to_cart("Chicken Wings", 1)
show_cart()

# 6. Remove Gulab Jamun
remove_from_cart("Gulab Jamun")
show_cart()


# Final Order Summary
print("\n========== Order Summary ==========")

subtotal = 0
for entry in cart:
    item = entry["item"]
    qty = entry["quantity"]
    price = entry["price"]

    total = qty * price
    subtotal += total

    print(f"{item:<18} x{qty}    ₹{total:.2f}")

print("------------------------------------")

gst = subtotal * 0.05
total_payable = subtotal + gst

print(f"Subtotal:                ₹{subtotal:.2f}")
print(f"GST (5%):                ₹{gst:.2f}")
print(f"Total Payable:           ₹{total_payable:.2f}")
print("====================================")
```

```python
# Task 3: Inventory Tracker with Deep Copy

import copy  # used for deep copying

# Create a deep copy of inventory

inventory_backup = copy.deepcopy(inventory)

# Demonstrate deep copy
print("Before making any changes:")
print("Original stock (Paneer Tikka):", inventory["Paneer Tikka"]["stock"])
print("Backup stock   (Paneer Tikka):", inventory_backup["Paneer Tikka"]["stock"])

# changing stock in original inventory
inventory["Paneer Tikka"]["stock"] = 5

print("\nAfter changing original inventory:")
print("Original stock (Paneer Tikka):", inventory["Paneer Tikka"]["stock"])
print("Backup stock   (Paneer Tikka):", inventory_backup["Paneer Tikka"]["stock"])

# restoring original value using backup
inventory["Paneer Tikka"]["stock"] = inventory_backup["Paneer Tikka"]["stock"]

print("\nAfter restoring original value:")
print("Original stock (Paneer Tikka):", inventory["Paneer Tikka"]["stock"])


# Deduct stock based on cart
print("\nUpdating inventory after order...")

for entry in cart:
    item_name = entry["item"]
    quantity = entry["quantity"]

    if item_name in inventory:
        current_stock = inventory[item_name]["stock"]

        # checking if enough stock is available
        if current_stock >= quantity:
            inventory[item_name]["stock"] -= quantity
        else:
            print(f"Warning: Only {current_stock} unit(s) of {item_name} available.")
            inventory[item_name]["stock"] = 0


# Reorder Alerts

print("\nReorder Alerts:")
found_alert = False

for item in inventory:
    stock = inventory[item]["stock"]
    reorder_level = inventory[item]["reorder_level"]

    if stock <= reorder_level:
        print(f"⚠ Reorder Alert: {item} — Only {stock} unit(s) left (reorder level: {reorder_level})")
        found_alert = True

if not found_alert:
    print("No items currently need reordering.")


# Final check

print("\nFinal Comparison:")

print("Current inventory (Paneer Tikka):", inventory["Paneer Tikka"])
print("Backup inventory  (Paneer Tikka):", inventory_backup["Paneer Tikka"])
```

```python
# Task 4: Daily Sales Log Analysis

# Total revenue per day
print("Revenue per day:")

daily_revenue = {}

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]

    daily_revenue[date] = total
    print(f"{date} : ₹{total}")


# Best-selling day
best_day = max(daily_revenue, key=daily_revenue.get)

print("\nBest Selling Day:")
print(f"{best_day} with revenue ₹{daily_revenue[best_day]}")


# Most ordered item
item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item in item_count:
                item_count[item] += 1
            else:
                item_count[item] = 1

# finding most ordered item
most_ordered = max(item_count, key=item_count.get)

print("\nMost Ordered Item:")
print(f"{most_ordered} ({item_count[most_ordered]} orders)")


# Add new day

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

print("\nAfter adding new day:\n")

# recompute revenue
daily_revenue = {}

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]

    daily_revenue[date] = total
    print(f"{date} : ₹{total}")

# best day again
best_day = max(daily_revenue, key=daily_revenue.get)

print("\nUpdated Best Selling Day:")
print(f"{best_day} with revenue ₹{daily_revenue[best_day]}")


# Print all orders using enumerate
print("\nAll Orders:\n")

count = 1

for date in sales_log:
    for order in sales_log[date]:
        items = ", ".join(order["items"])
        print(f"{count}. [{date}] Order #{order['order_id']} — ₹{order['total']} — Items: {items}")
        count += 1
```
